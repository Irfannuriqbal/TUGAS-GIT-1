```python
data_panen = {
    'lokasi1': {
        'nama_lokasi': 'Kebun A',
        'hasil_panen': {
            'padi': 1200,
            'jagung': 800,
            'kedelai': 500
        }
    },
    'lokasi2': {
        'nama_lokasi': 'Kebun B',
        'hasil_panen': {
            'padi': 1500,
            'jagung': 900,
            'kedelai': 450
        }
    },
    'lokasi3': {
        'nama_lokasi': 'Kebun C',
        'hasil_panen': {
            'padi': 1100,
            'jagung': 750,
            'kedelai': 600
        }
    },
    'lokasi4': {
        'nama_lokasi': 'Kebun D',
        'hasil_panen': {
            'padi': 1300,
            'jagung': 850,
            'kedelai': 550
        }
    },
    'lokasi5': {
        'nama_lokasi': 'Kebun E',
        'hasil_panen': {
            'padi': 1400,
            'jagung': 950,
            'kedelai': 480
        }
    }
}

def tampilkan_detail_panen(data):
    """Menampilkan detail panen untuk setiap lokasi"""
    print(">>> Laporan Detail Panen <<<")
    for kode, info in data.items():
        print(f"Lokasi: {info['nama_lokasi']} ({kode})")
        for komoditas, jumlah in info['hasil_panen'].items():
            print(f" - {komoditas.upper()}: {jumlah} kg")
        print("----------------------------")

def analisis_panen(data):
    """Menganalisis hasil panen berdasarkan kriteria tertentu"""
    print("\n>>> Analisis Kondisi Panen <<<")
    padi_per_lokasi = {}
    kedelai_per_lokasi = {}

    for kode, info in data.items():
        padi = info['hasil_panen']['padi']
        jagung = info['hasil_panen']['jagung']
        kedelai = info['hasil_panen']['kedelai']

        padi_per_lokasi[kode] = padi
        kedelai_per_lokasi[kode] = kedelai

        print(f"Evaluasi {info['nama_lokasi']}:")
        print(f"  - Padi: {padi} kg")
        print(f"  - Jagung: {jagung} kg")
        print(f"  - Kedelai: {kedelai} kg")

        if padi > 1300 or jagung > 800:
            print("  - Status: PK Butuh Perhatian Khusus!")
        else:
            print("  - Status: KN Kondisi Normal")
        print("--" * 30)

    return padi_per_lokasi, kedelai_per_lokasi

#Eksekusi fungsi-fungsi
tampilkan_detail_panen(data_panen)
padi_lokasi, kedelai_lokasi = analisis_panen(data_panen)

# Informasi spesifik
print("\n>>> Informasi Khusus <<<")
print(f"Hasil Jagung Lokasi 2: {data_panen['lokasi2']['hasil_panen']['jagung']} kg")
print(f"Nama Lokasi 3: {data_panen['lokasi3']['nama_lokasi']}")

print("\n>>> Rekap Padi per Lokasi <<<")
for lokasi, jumlah in padi_lokasi.items():
    print(f"{lokasi}: {jumlah} kg")

print("\n>>> Rekap Kedelai per Lokasi <<<")
for lokasi, jumlah in kedelai_lokasi.items():
    print(f"{lokasi}: {jumlah} kg")
```
